```python
import torch
import torch.nn as nn
import torch.optim as optim

class DNN(torch.nn.Module):
    def __init__(self, input_size = 40, hidden_size = 128, output_size = 8):
        super(DNN, self).__init__()
        self.hidden = torch.nn.Linear(input_size, hidden_size)
        self.relu = torch.nn.ReLU()
        self.output = torch.nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.hidden(x)
        x = self.relu(x)
        x = self.output(x)
        return x

#load data
#I used jupyter to test this so that I run get_datat.ipynb here
#not recommanded to use jupyter for this project 
%run get_data.ipynb
train_data = train_dataset.to("cuda")
train_labels = label_dataset.to("cuda")
train_num = len(train_data)

#If you use vscode or pycharm, pls use the following block code
'''
import get_data
train_data = get_data.train_dataset.to("cuda")
train_labels = get_data.label_dataset.to("cuda")
train_num = 1440
'''

#Initialize the Model and Optimizer
model = DNN().to("cuda")
optimizer = optim.Adam(model.parameters(), lr=1E-4)
criterion = nn.CrossEntropyLoss()

#train
for epoch in range(1024):    #choose the num of epoch as you want
    optimizer.zero_grad()
    outputs = model(train_data)

    loss = criterion(outputs, train_labels)
    loss.backward()
    optimizer.step()

    if (epoch+ 1)%10 == 0:
        accuracy = (outputs.argmax(1) == train_labels).type(torch.float32).sum().item() / train_num
        print("epochï¼š",epoch,"train_loss:", (loss),"accuracy:",(accuracy))
        print("-------------")

#save the trained model
model_save_path = '/path/for/the/emotion_model.pth'  
torch.save(model.state_dict(), model_save_path)
print(f"Model saved as {model_save_path}")
```
